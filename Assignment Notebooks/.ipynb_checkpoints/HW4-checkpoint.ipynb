```python
import sys
from matplotlib import pyplot as plt
import numpy as np
import astropy.units as u
from scipy.special import expn, exp1
from scipy.integrate import simps 
import math

import warnings

# Ignore all warnings
warnings.filterwarnings("ignore")

# importing custom functions

sys.path.append("../")
from ASTRO530.functions import integrate_bounds, precision, plot_precisions_bdx, plot_precisions_abdx
from ASTRO530 import plotting
```

```python
def H_nu_tau(tau,a_n):
    b = 10e2 # limits set from HW3 for maximum precision
    a = 10e-18
    dx = 10e-5

    if tau != 0.0 and tau > a:
        num = int(round((np.log10(b)-np.log10(tau))/dx,0))
        t = np.logspace(np.log10(tau),np.log10(b),num)
        term1 = 0.5*simps(S_nu(t,a_n)*expn(2,t-tau),x=t) # integrate from tau to infty
        
        num = int(round(((np.log10(tau)-np.log10(a))/dx),0))
        t = np.logspace(np.log10(a),np.log10(tau),num)
        t[0] = 0
        term2 = -0.5*simps(S_nu(t,a_n)*expn(2,tau-t),x=t) # integrate from 0 to tau
        H = term1 + term2

    else:
        num = int(round(b/dx,0))
        t = np.linspace(tau,b,num)
        H = 0.5*simps(S_nu(t,a_n)*expn(2,t),x=t)

    return H     
```

```python
def S_nu(tau,a_n): # source function, n is the order of terms
    sum = 0
    for i, a in enumerate(a_n):
        sum += a*tau**i
    return sum

def H_nu_plot(tau_array,S_nu_order):

    a_n = [1]*(S_nu_order+1) # only having a_n coeffients be = 1

    b = 10e2 # limits set from HW3 for maximum precision
    a = 10e-18
    dx = 10e-5

    H_array = []
    for i,tau in enumerate(tau_array): 
        H_array.append(H_nu_tau(tau,a_n))

    H = H_array[-1]
    # tau_0 = tau_array[np.where(H_array == np.min(H_array))]
    tau_0 = tau_array[np.where(H_array == H_array[-1])]
    # print("The lowest value of H_nu = ", H, "occurred at tau_nu =", tau_0)

    plt.plot(tau_array,H_array,lw=4,c=['navy','salmon'][S_nu_order-1],label=[r'Linear $S_{\nu}$',r'Quadratic $S_{\nu}$'][S_nu_order-1],zorder=0)
    plt.scatter(tau_0,H,s=200,zorder=1,c=['navy','salmon'][S_nu_order-1],label=r'$H_{\nu} (\tau_{\nu}=0)$')
    
    return H

fig, ax = plt.subplots(figsize=(12,8))

tau_test = np.logspace(0,-5,25)
tau_test[-1] = 0
H_nu_plot(tau_test,2)  # quadratic source function  
H_nu_plot(tau_test,1)  # linear source function

ax.set_ylabel(r'$H_{\nu} (\tau_{\nu})$ [$\mathrm{erg} \ s^{-1} \ cm^{-1} \ Hz^{-1}$]')
ax.set_xlabel(r'$\tau_{\nu}$')
ax.grid()
plt.legend()
plt.savefig("../Plots/HW4/H_nu.pdf")
plt.show()
```
